```python
np.random.seed(42)
df3 = pd.DataFrame(  np.random.randn(5,4), 
                   index = ['A', 'B', 'C', 'D', 'E'], 
                   columns = 'W X Y Z'.split()          )
df3
```

```python
df6 = pd.DataFrame({'groups': ['A', 'B', 'C', 'A', 'B', 'C','A', 'B', 'C'],
                   'var1': [10,23,33,22,11,99,40,20,15],
                   'var2': [100,253,333,262,111,969, 232,343,123]})
df6
```

```python
df6[["var1"]].transform(lambda x : ( x - df6["var1"].min() )/ 
                                       (df6["var1"].max()-df6["var1"].min())
                       )  # make minmaxscaler
```

```python
df6[["var1","var2"]].transform(lambda x : (x-x.min())
                               /(x.max()-x.min()))
                       
```

```python
df6[["var2"]].apply(lambda x:(x-df6["var2"].min())/ 
                    (df6["var2"].max()-df6["var2"].min()) )  # make minmaxscaler
```

```python
df6[["var1","var2"]].apply(lambda x : (x-x.min())
                               /(x.max()-x.min()))
                       
```

```python
df6.groupby('groups')[["var1","var2"]].transform(np.mean)
```
